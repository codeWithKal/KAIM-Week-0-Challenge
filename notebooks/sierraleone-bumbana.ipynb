# Imporing and exploring the dataset of sierraleone-bumbuna.csv

```python
df_srln = pd.read_csv("../src/csv-files/sierraleone-bumbuna.csv")
```

```python
df_srln.head()
```

```python
df_srln.info()
```

```python
# From the above cell we see that Comments column is empty and thus should be dropped
df_srln.drop(columns="Comments", inplace=True)
df_srln.info()
```

```python
df_srln.describe()
```

```python
# Lemme now check if there is any null value there among the values of each column
df_srln.isnull().sum()
```

let me now change the timestamp column to be date object and make it an index of the new data frame

```python
df_srln["Timestamp"] = pd.to_datetime(df_srln["Timestamp"])
df_srln.set_index("Timestamp", inplace=True)
df_srln.head()
```

```python
len(df_srln)
```

Now it appears that the data is taken within difference of minuts lets resample it to frame of 1H

```python
df_srln.shape
```

```python
df_srln = df_srln.resample('h').mean()
df_srln.head()
```

```python
df_srln.shape
```

Now the number of our record is reduced significantly from (525600, 17) to (8761, 17)

```python
df_srln.describe()
```

Across the cleaned dataframe we see that only 5 columns help to deal with our business. i'e GHI, DNI, DHI, ModA, ModB.

```python
# Lemme create a clean data frame that takes the mentioned columns from the df_srln dataframe.
columns_to_keep = ["GHI", "DNI", "DHI", "ModA", "ModB"]
df = df_srln[columns_to_keep]
df.head()
```

```python
df.describe()
```

```python
df["GHI"].plot(kind="box")
plt.title("Frequency Distribution for GHI")
plt.ylabel("Frequency")
```

From the above plotbox we see that ther is no outliers for the GHI column

```python
df.head()
```

```python
# Let me now do the same thing to the rest four columns and check for any outlier.
df.plot(kind="box")
```

```python
# from the above plot box we can see that DNI and DHI have outliers and hence should be cleaned.
df.describe()
```

```python
df.describe()
```

```python
low, high = df["DNI"].quantile(0.1), df["DNI"].quantile(0.9)
df = df[df["DNI"].between(low, high)]
df.head()
```

```python
mean_benin_GHI = df["GHI"].mean()
mean_benin_DNI = df["DNI"].mean()
mean_benin_DHI= df["DHI"].mean()
print(mean_benin_GHI, mean_benin_DNI, mean_benin_DHI);
```
