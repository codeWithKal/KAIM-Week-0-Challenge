# Imporing and exploring the dataset of togo-depaong_qc.csv

```python
df_togo = pd.read_csv("../src/csv-files/togo-dapaong_qc.csv")
```

```python
df_togo.head()
```

```python
df_togo.info()
```

```python
# From the above cell we see that Comments column is empty and thus should be dropped
df_togo.drop(columns="Comments", inplace=True)
df_togo.info()
```

```python
df_togo.describe()
```

```python
# Lemme now check if there is any null value there among the values of each column
df_togo.isnull().sum()
```

let me now change the timestamp column to be date object and make it an index of the new data frame

```python
df_togo["Timestamp"] = pd.to_datetime(df_togo["Timestamp"])
df_togo.set_index("Timestamp", inplace=True)
df_togo.head()
```

```python
len(df_togo)
```

Now it appears that the data is taken within difference of minuts lets resample it to frame of 1H

```python
df_togo.shape
```

```python
df_togo = df_togo.resample('h').mean()
df_togo.head()
```

```python
df_togo.shape
```

Now the number of our record is reduced significantly from (525600, 17) to (8761, 17)

```python
df_togo.describe()
```

Across the cleaned dataframe we see that only 5 columns help to deal with our business. i'e GHI, DNI, DHI, ModA, ModB.

```python
# Lemme create a clean data frame that takes the mentioned columns from the df_togo dataframe.
columns_to_keep = ["GHI", "DNI", "DHI", "ModA", "ModB"]
df_togo_cleaned = df_togo[columns_to_keep]
df_togo_cleaned.head()
```

```python
df_togo_cleaned.describe()
```

```python
df_togo_cleaned["GHI"].plot(kind="box")
plt.title("Frequency Distribution for GHI")
plt.ylabel("Frequency")
```

From the above plotbox we see that ther is no outliers for the GHI column

```python
df_togo_cleaned.head()
```

```python
# Let me now do the same thing to the rest four columns and check for any outlier.
df_togo_cleaned.plot(kind="box")
```

```python
# from the above plot box we can see that DNI and DHI have outliers and hence should be cleaned.
df_togo_cleaned.describe()
```
