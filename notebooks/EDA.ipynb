## Let's now see the mean GHI, mean DNI and mean DHI measures of three places.


```python
GHI_mean = [mean_benin_GHI, mean_srln_GHI, mean_togo_GHI]
GHI_mean_identifier = ["mean_benin_GHI", "mean_srln_GHI", "mean_togo_GHI"]
plt.bar(GHI_mean_identifier, GHI_mean)
plt.ylabel("Mean")
plt.title("Mean barchart for the GHI measure of the three places");
plt.savefig(save_path + "GHI_comparison.png")
```

```python
DHI_mean = [mean_benin_DHI, mean_srln_DHI, mean_togo_DHI]
DHI_mean_identifier = ["mean_benin_DHI", "mean_srln_DHI", "mean_togo_DHI"]
plt.bar(DHI_mean_identifier, DHI_mean)
plt.ylabel("Mean")
plt.title("Mean barchart for the DHI measure of the three places");
plt.savefig(save_path + "DHI_comparison.png")
```

```python
DNI_mean = [mean_benin_DNI, mean_srln_DNI, mean_togo_DNI]
DNI_mean_identifier = ["mean_benin_DNI", "mean_srln_DNI", "mean_togo_DNI"]
plt.bar(DNI_mean_identifier, DNI_mean)
plt.ylabel("Mean")
plt.title("Mean barchart for the DNI measure of the three places");
plt.savefig(save_path+"DNI-comparison.png")
```

```python
places = ["benin-malanville", "sierraleone-bumbana", "togo-dapaong"]
n = len(places);
x= np.arange(n)
width = 0.25
plt.figure(figsize=(10, 6))
plt.bar(x - width, GHI_mean, width, label="GHI-mean")
plt.bar(x, DHI_mean, width, label="DHI-mean")
plt.bar(x + width, DNI_mean, width, label="DNI-mean")
plt.title("aggregate mean values of GHI, DHI, DNI")
plt.xlabel("Site Places")
plt.ylabel("Mean-values")
plt.xticks(x, places)
plt.legend()
plt.savefig(save_path+ "all_barcharts.png")
```
