# Imporing and exploring the dataset of benin-malanville.csv

```python
# From the above cell we see that Comments column is empty and thus should be dropped
df_benin.drop(columns="Comments", inplace=True)
df_benin.info()
```

```python
df_benin.describe()
```

```python
# Lemme now check if there is any null value there among the values of each column
df_benin.isnull().sum()
```

let me now change the timestamp column to be date object and make it an index of the new data frame

```python
df_benin["Timestamp"] = pd.to_datetime(df_benin["Timestamp"])
df_benin.set_index("Timestamp", inplace=True)
df_benin.head()
```

```python
len(df_benin)
```

Now it appears that the data is taken within difference of minuts lets resample it to frame of 1H

```python
df_benin.shape
```

```python
df_benin = df_benin.resample('h').mean()
df_benin.head()
```

```python
df_benin.shape
```
